# Data Visualization II
1. Use the inbuilt dataset 'titanic' as used in the above problem. Plot a box plot for distribution
of age with respect to each gender along with the information about whether they survived
or not. (Column names : 'sex' and 'age')
2. Write observations on the inference from the above statistics

```python
# Step 1: Import Libraries and Load Data

import seaborn as sns
import matplotlib.pyplot as plt

# Load the inbuilt titanic dataset
titanic = sns.load_dataset('titanic')

# Display the first few rows
print("Dataset Loaded Successfully.")
display(titanic.head())
```

```python
# Step 2: Statistical Details using describe()

print("\n--- Statistical Summary of Age and Survival ---")
# We filter for only relevant columns to keep the output clean
display(titanic[['age', 'survived']].describe())

# Check for missing values in Age (common in this dataset)
print("\nMissing values in Age:", titanic['age'].isnull().sum())
```

```python
# Step 3: Create the Box Plot

# Set the plot size and style
plt.figure(figsize=(12, 7))
sns.set_style("whitegrid")

# Create the box plot
sns.boxplot(x='sex', y='age', hue='survived', data=titanic, palette='Set1')

# Adding labels and title
plt.title('Distribution of Age with respect to Gender and Survival Status', fontsize=15)
plt.xlabel('Gender (Sex)', fontsize=12)
plt.ylabel('Age', fontsize=12)
plt.legend(title='Survived', labels=['No', 'Yes'])

plt.show()
```

Step 4:
Observations and Inference
Based on the box plot generated, here are the key statistical observations:

Survival of the Young: For both males and females, the median age of those who survived (Yes) is generally lower or shows a wider spread in the lower quartiles compared to those who did not. This aligns with the "children first" evacuation policy.

Gender Difference: The survival rate for females is much higher regardless of age, but the box plot shows that elderly females had a significantly higher survival rate than elderly males.

Male Survival Outliers: There are several outliers in the male category (dots above the whiskers). This indicates that while most older men did not survive, a few very elderly men (aged 60-80) were among the survivors.

Age Spread: The age spread for non-surviving males is concentrated between 20 and 40 years, indicating that many young-to-middle-aged men did not make it off the ship.

Data Gaps: The whiskers and box size tell us that the "Survived" group for males has a smaller interquartile range (IQR) than the "Not Survived" group, meaning surviving males were mostly within a specific younger age bracket.
