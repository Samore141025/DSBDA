Descriptive Statistics - Measures of Central Tendency and variability
Perform the following operations on any open source dataset

Write a Python program to display some basic statistical details like percentile, mean,
standard deviation etc. of the species of ‘Iris-setosa’, ‘Iris-versicolor’ and ‘Iris-versicolor’
of iris.csv dataset.

```python
import pandas as pd

# Load the dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

# Display the first few rows to verify loading
print("Dataset Head:")
print(df.head())
```

```python
# Group the data by 'species' and calculate descriptive statistics
# The describe() function automatically provides mean, std, min, 25%, 50%, 75%, and max
stats = df.groupby('species').describe()

# Display the statistics
print("\nStatistical Details by Species:")
print(stats)
```

```python
# List of species to analyze
species_list = ['setosa', 'versicolor', 'virginica']

for species in species_list:
    print(f"\n--- Statistics for Iris-{species} ---")
    # Filter the dataframe and describe
    species_stats = df[df['species'] == species].describe()
    print(species_stats)
```
