1. Use the inbuilt dataset 'titanic'. The dataset contains 891 rows and contains information
about the passengers who boarded the unfortunate Titanic ship. Use the Seaborn library to
see if we can find any patterns in the data.
2. Write a code to check how the price of the ticket (column name: 'fare') for each passenger
is distributed by plotting a histogram.


```python
# Step 1: Import Libraries and Load the Inbuilt Dataset

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the inbuilt titanic dataset
titanic = sns.load_dataset('titanic')

# Display the first few rows
print("Titanic Dataset Preview:")
display(titanic.head())
```

```python
# Step 2: Basic Exploration with describe()

print("\n--- Statistical Summary of Titanic Dataset ---")
display(titanic.describe())

# Checking for missing values
print("\nMissing values in each column:")
print(titanic.isnull().sum())
```

```python
# Step 3: Distribution of Ticket Prices (Histogram)

# Set the visual style
sns.set_style('whitegrid')

# Plotting the histogram for 'fare'
plt.figure(figsize=(10, 6))
sns.histplot(titanic['fare'], bins=30, kde=True, color='blue')

plt.title('Distribution of Ticket Prices (Fare)')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
# Step 4: Finding Patterns (Relationships)

# 4.A Survival Rate by Gender:

# Bar plot to see survival patterns by sex
plt.figure(figsize=(8, 5))
sns.barplot(x='sex', y='survived', data=titanic)
plt.title('Survival Rate by Gender')
plt.show()

# 4.B Survival Rate by Passenger Class:
# Count plot to see survival patterns by class
plt.figure(figsize=(8, 5))
sns.countplot(x='pclass', hue='survived', data=titanic)
plt.title('Survival Count by Passenger Class')
plt.show()
```
